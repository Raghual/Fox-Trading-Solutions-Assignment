```python
#Importing necessary libraries
import xlwings as xw
import requests, json 
import time
```

```python
#Assigning API to api_key variable
api_key = "My API"
```

```python
#Base Url of Openweather Website
base_url = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
#Opening the Blank Excel File stored locally in my system.
wbTest=xw.Book(r'C:\Users\K\xlautomation\Output_file.xlsx')
```

```python
#Creating Instances of Sheets(I have changed the name of sheets as 'Tab-1','Tab-2',Tab-3)
ws1=wbTest.sheets['Tab-1']
ws2=wbTest.sheets['Tab-2']
ws3=wbTest.sheets['Tab-3']
```

```python
ws1.cells(1,"A").value="City"
ws1.cells(1,"B").value="Temperature"
ws1.cells(1,"C").value="Humidity"
ws1.cells(1,"D").value="Unit"
ws1.cells(1,"E").value="Update"
city_ctr,temp_ctr,hum_ctr,up_ctr,cf_ctr=2,2,2,2,2
```

```python
No_of_cities_input=int(input())
```

```python
#This block will get the necessary inputs,send the request to the server, get the response from the server, extract the necessary keys from the returned json output, and store the data in excel.
while(No_of_cities_input>0):
    city_name = input()
    C_or_F=input()
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name
    update=int(input())
    ws1.cells(city_ctr,"A").value=city_name
    ws1.cells(city_ctr,"D").value=C_or_F
    ws1.cells(up_ctr,"E").value=update
    if(update==1):
        times=60
        while(times>0):
          response = requests.get(complete_url) 
          x = response.json() 
          if x["cod"] != "404": 
              y = x["main"] 
              current_temperature = y["temp"] 
              current_humidity = y["humidity"] 
              if(C_or_F=="C"):
                current_temperature=str(current_temperature-273.15)
              else:
                current_temperature=str((current_temperature-273.15)*1.8+32)
              print(" Temperature = ",current_temperature,"\n"," Humidity = ",current_humidity)
              ws1.cells(temp_ctr,"B").value=current_temperature
              ws1.cells(hum_ctr,"C").value=current_humidity
              time.sleep(1)
              times-=1

        
    else:
        response = requests.get(complete_url) 
        x = response.json() 
        if x["cod"] != "404": 
              y = x["main"] 
              current_temperature = y["temp"] 
              current_humidity = y["humidity"] 
              if(C_or_F=="C"):
                current_temperature=str(current_temperature-273.15)
              else:
                current_temperature=str((current_temperature-273.15)*1.8+32)
              print(" Temperature = ",current_temperature,"\n"," Humidity = ",current_humidity)
              ws1.cells(temp_ctr,"B").value=current_temperature
              ws1.cells(hum_ctr,"C").value=current_humidity
    city_ctr+=1
    temp_ctr+=1
    hum_ctr+=1
    up_ctr+=1
    cf_ctr+=1
    No_of_cities_input-=1
    
```
